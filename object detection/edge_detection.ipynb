# Edge Detection
* 캐니 가장자리 검출
  * 노이즈를 제거하기 위해 가우스 필터를 적용하여 이미지 적용을 부드럽게함 (스무딩 블러딩 기술)
  * 약한 모서리를 없애고 강한 모서리만을 감지함
  * 사용자가 임계값을 지정함

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('./DATA/sammy_face.jpg')
```

```python
plt.imshow(img)
```

```python
edges = cv2.Canny(image=img, threshold1=127, threshold2=127)
```

```python
plt.imshow(edges)
```

```python
edges = cv2.Canny(image=img, threshold1=0, threshold2=225)
plt.imshow(edges)
```

```python
med_val = np.median(img)
```

```python
med_val
```

```python
#하한 임계값을 0 또는 중앙의 70%로 설정 
lower = int(max(0, 0.7*med_val))
#상향 임계값을 130% 또는 255로 설정
upper = int(min(255, 1.3*med_val))
```

```python
edges = cv2.Canny(image=img, threshold1=lower, threshold2=upper)
plt.imshow(edges)
```

```python
upper
```

```python
edges = cv2.Canny(image=img, threshold1=lower, threshold2=upper+50)
plt.imshow(edges)
```

```python
blurred_img = cv2.blur(img, ksize=(5,5))
```

```python
edges = cv2.Canny(image=blurred_img, threshold1=lower, threshold2=upper+50)
plt.imshow(edges)
```
